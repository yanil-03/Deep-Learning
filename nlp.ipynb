```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text':['people watch cricket','player was player',
                           'people write comment',
                           'cricket play comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(['cricket watch and write comment of cricket comment']).toarray()
```

```python
# 2. N-grams
v = CountVectorizer(ngram_range=(1,2))
```

```python
bow = v.fit_transform(df['text'])
```

```python
df['text']
```

```python
print(v.vocabulary_)
```

```python
# TF(term frequency)
# 2. IDF(inverse document frequency)
```

```python
# from ast import MatchSingleton
# TF(t,d) = (number of occurance of term t in document d)/(total number of terms in document d)
# example=>
# d1 = people watch cricket

# tf(people,d1) = 1/3

# d2 = match was match
# tf(match,d2) = 2/3
```

```python
# from logging import log
# IDF(t) = Loge(Total number of documents)/(number of documents with term t in it)

# here corpus means number of sentences

# Example ==>
# d1 = people watch cricket
# d2 = cricket was cricket
# d3 = people play comment
# d4 = cricket write comment

# idf(cricket) = log(4/4)
# idf(people) = log(4/2)
```

```python
# we can use this by apply feature_extraction method in sklearn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
