```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files
path = list(files.upload().keys())[0]
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.sample(5000)
```

```python
# remove punctuations
import string
punct = string.punctuation
punct
```

```python
def remove_punct(text):
  my_list = []
  for i in text.split():
    if i not in punct:
      my_list.append(i)
  return ' '.join(my_list)
```

```python
df['Job Title'] = df['Job Title'].apply(remove_punct)
```

```python
df['Location'] = df['Location'].apply(remove_punct)
```

```python
df['Company Name'] = df['Company Name'].astype(str)
```

```python
df['Company Name'] = df['Company Name'].apply(remove_punct)
```

```python
df['Content'] = df.apply(lambda x : ' '.join(x.dropna().astype(str)) , axis = 1)
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000)
```

```python
metrics = cv.fit_transform(df['Content']).toarray()
```

```python
metrics.shape
```

```python
similarity = cosine_similarity(metrics)
```

```python
def recommend(job):
  # Get the index from the sampled dataframe
  index = df[df['Job Title'] == job].index
  if len(index) == 0:
    return "Job title not found in the sampled data."
  index = index[0]

  # Map the original index to the new index in the sampled dataframe
  sampled_index = df.reset_index().index[df.reset_index()['index'] == index].values[0]

  ditances = sorted(enumerate(similarity[sampled_index]), reverse = True, key = lambda x:x[1])
  new_index = [i[0]for i in ditances[1:11]]
  return df.iloc[new_index, :-1]
```

```python
df.sample(5)
```

```python
recommend('Director')
```

```python
recommend('Web Developer')
```
