```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("newplacementdata.csv")
df.head()
```

```python
sns.scatterplot(x="cgpa",y="placement_exam_marks", data = df,hue="placed")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = ["placed"])
y = df["placed"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Perceptron
pr = Perceptron()
```

```python
pr.fit(x_train, y_train)
```

```python
pr.score(x_train, y_train)*100 , pr.score(x_test, y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
```
